```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3'],})
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7'],})
```

```python
#para este caso los indices heredan el anterior
#por default esta concatenacion lo une por el axis 0 que es el de columnas
pd.concat([df1,df2])
#se corrije asi
pd.concat([df1,df2],ignore_index = True)
```

```python
#Ahora se a√±ade por columnas usando axis 1
pd.concat([df1,df2], axis = 1)
```

## Merge

```python
left = pd.DataFrame({'key': ['k0','k1','k2','k3'],
 'A': ['A0','A1','A2','A3'],
 'B': ['B0','B1','B2','B3'],
 })
right = pd.DataFrame({'key': ['k0','k1','k2','k3'],
 'C': ['C0','C1','C2','C3'],
 'D': ['D0','D1','D2','D3'],
 })
```

```python
# Sintaxis para unir dos bases de datos por medio de un indice
# dfleft.merge(dfright, on = 'indice')
left.merge(right, on = 'key')
```

```python
left = pd.DataFrame({'key': ['k0','k1','k2','k3'],
 'A': ['A0','A1','A2','A3'],
 'B': ['B0','B1','B2','B3'],
 })
right = pd.DataFrame({'key2': ['k0','k1','k2','k3'],
 'C': ['C0','C1','C2','C3'],
 'D': ['D0','D1','D2','D3'],
 })
```

```python
left.merge(right, left_on = 'key', right_on= 'key2')
```

```python
left = pd.DataFrame({'key': ['k0','k1','k2','k3'],
 'A': ['A0','A1','A2','A3'],
 'B': ['B0','B1','B2','B3'],
 })
right = pd.DataFrame({'key2': ['k0','k1','k2',np.nan],
 'C': ['C0','C1','C2','C3'],
 'D': ['D0','D1','D2','D3'],
 })
right
```

```python
#No trae los valores que no coinciden
left.merge(right, left_on = 'key', right_on= 'key2')
```

```python
#se soluciona con how
left.merge(right, left_on = 'key', right_on= 'key2', how = 'left')
```

```python
#se soluciona con how
left.merge(right, left_on = 'key', right_on= 'key2', how = 'right')
```

## Join index match

```python
left = pd.DataFrame({'A': ['A0','A1','A2'],
  'B':['B0','B1','B2']},
  index=['k0','k1','k2'])

right =pd.DataFrame({'C': ['C0','C1','C2'],
  'D':['D0','D1','D2']},
  index=['k0','k2','k3']) 
```

```python
# Hace una operacion de join definida por los indices
left.join(right)
```

```python
left.join(right, how = 'inner')
```

```python
left.join(right, how = 'right')
```

```python
left.join(right, how = 'left')
```

```python
left.join(right, how = 'outer')
```
