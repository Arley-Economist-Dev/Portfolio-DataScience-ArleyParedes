## modificaciones tipo matlab

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1, 'r--')
ax.plot(x,x+2, 'g-')
ax.plot(x,x+3, 'y.-')
ax.plot(x,x+4, 'bx:')
```

## Pyplot costumization

```python
fig, ax = plt.subplots(figsize=(8,8))

#se puede configurar un alpha para la transparencias
ax.plot(x,x+1, color = 'green', alpha = 0.8, linewidth = 5, linestyle = '-', 
        marker='o', markersize = 10, markerfacecolor = '#CCFFCC')
ax.plot(x,x+2, color = 'red', linewidth = 4, linestyle = '--', marker='x')
#admite formato rgb
ax.plot(x,x+3, color = '#66FF66', linewidth = 3, linestyle = 'dashed', marker='v')
ax.plot(x,x+4, color = '#CCFFCC', linewidth = 2, linestyle = ':', marker='P')
```
