## dependencias
Esto sirve para graficar variables categoricas, es decir de texto.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPON', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 400, 340]
```

```python
plt.bar(country, population, width = 0.5, color = ('red', 'blue'))
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), 
           rotation = 45)
plt.show()
```

```python
plt.barh(country, population)
plt.show()
```
