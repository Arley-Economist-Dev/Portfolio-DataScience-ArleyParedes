## Dependencias

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = False, hue = 'sex', 
             stat = 'frequency')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = False, hue = 'sex', 
             stat = 'probability')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = False, hue = 'sex', 
             stat = 'percent')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = False, hue = 'sex', 
             stat = 'density')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = False, hue = 'sex', 
             stat = 'count')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = False, hue = 'sex', 
             stat = 'count', multiple = 'stack')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = False, hue = 'sex', 
             stat = 'count', multiple = 'layer')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = False, hue = 'sex', 
             stat = 'count', multiple = 'dodge')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative = False, hue = 'sex', 
             stat = 'count', multiple = 'fill')
plt.show()
```

```python
# Densidad de Kernel
sns.kdeplot(data = tips,x = 'tip', hue = 'sex', cumulative = False,
            shade = True, bw_adjust = 1)
plt.show()
```

```python
sns.ecdfplot(data = tips, x = 'tip', hue = 'sex', stat = 'count')
plt.show()
```

```python
sns.displot(data = tips, x = 'tip', hue = 'sex', kind = 'kde')
plt.show()
```

```python
sns.displot(data = tips, x = 'tip', hue = 'sex', kind = 'hist')
plt.show()
```

```python
sns.displot(data = tips, x = 'tip', hue = 'sex', kind = 'hist',
            multiple = 'stack')
plt.show()
```
