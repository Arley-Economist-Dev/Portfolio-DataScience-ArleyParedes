## Dependencias

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.countplot(data=tips, x = 'day', hue = 'sex')
plt.show()
```

```python
sns.countplot(data=tips, y = 'day', hue = 'sex')
plt.show()
```

```python
sns.stripplot(data=tips, x = 'day', y = 'total_bill',  hue = 'sex',
              dodge = True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x = 'day', y = 'total_bill',  hue = 'sex',
              dodge = True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x = 'day', y = 'total_bill',  hue = 'sex',
              dodge = True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x = 'day', y = 'total_bill',  hue = 'sex',
              dodge = True)
plt.show()
```

```python
sns.boxplot(data=tips, y = 'total_bill')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.boxplot(data=tips, x = 'day', y = 'total_bill',  hue = 'sex',
              dodge = True)
sns.swarmplot(data=tips, x = 'day', y = 'total_bill',  hue = 'sex',
              dodge = True, color = '0', marker = '<')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.violinplot(data=tips, x = 'day', y = 'total_bill',  hue = 'sex',
              dodge = True, palette = 'pastel', split = True)
plt.show()
```

```python
sns.violinplot(data = tips, y = 'total_bill')
```

```python
sns.catplot(data=tips, x = 'day', y = 'total_bill',  hue = 'sex',
            dodge = True, kind = 'swarm')
plt.show()
```

```python
sns.catplot(data=tips, x = 'day', y = 'total_bill',  hue = 'sex',
            dodge = True, kind = 'box', col='time')
plt.show()
```
